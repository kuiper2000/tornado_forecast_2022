```python
# environment setup
import os
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset as NetCDFFile
```

```python
# using command line to run download.sh 
# and put those data with python script under the same folder
```

```python
# dimension = target month (Jan-Dec), year (1992-2021), lat, lon 
data = np.load('SPEAR_ATS_ensmean_cape.npz')
cape = data['cape_SPEAR']

data = np.load('SPEAR_ATS_ensmean_z500.npz')
z500 = data['z500_SPEAR']

data = np.load('SPEAR_ATS_ensmean_WSHR.npz')
WSHR = data['WSHR_SPEAR']

data = np.load('SPEAR_ATS_ensmean_z925.npz')
z925 = data['z925_SPEAR']

# ivt share the same lon and lat with sst data
data = np.load('SPEAR_monthly_data_ivt_ensmean.npz')
ivtu = data['ivtu_model_ens']
ivtv = data['ivtv_model_ens']
```

```python
sst_model_ens      = np.zeros((15,360,360,576)) 
# load SST data
for init_time in ['01','02','03','04','05','06','07','08','09','10','11','12']: #initialization month
    for ens in range(1,16,1): # ensembler member
        nc    = NetCDFFile("SST_"+init_time+"_ens%02d.nc" % (ens,))
        data  = nc.variables['SST'][:,:,:]
        sst_model_ens[ens-1,:,:,:] = data[:360,:,:]

lat2    = nc.variables['lat']
lon2    = nc.variables['lon']
xx2,yy2 = np.meshgrid(lon2,lat2)
    
```
